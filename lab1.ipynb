### Normalizacja i podział daych

```python
diabetes = diabetes.dropna()
diabetes = diabetes.drop_duplicates()
```

```python
from sklearn.model_selection import train_test_split

# Stratified split is used because the dataset is imbalanced,
# ensuring that class proportions remain the same in train and test sets.
diabetes_train, diabetes_test = train_test_split(diabetes, test_size=0.2, random_state=1, stratify=diabetes["Diabetes_012"])

X_train = diabetes_train.drop("Diabetes_012", axis=1)
y_train = diabetes_train["Diabetes_012"]

X_test = diabetes_test.drop("Diabetes_012", axis=1)
y_test = diabetes_test["Diabetes_012"]

diabetes_train["Diabetes_012"].value_counts()
```

```python
diabetes_test["Diabetes_012"].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import pandas as pd

train_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
train_df['Diabetes_012'] = y_train.values

print("Original class distribution in train:")
print(train_df['Diabetes_012'].value_counts())
```

```python
min_count = train_df['Diabetes_012'].value_counts().min()

balanced_df = pd.concat([
    train_df[train_df['Diabetes_012'] == cls].sample(min_count, random_state=42)
    for cls in train_df['Diabetes_012'].unique()
])

balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

X_train_balanced = balanced_df.drop('Diabetes_012', axis=1)
y_train_balanced = balanced_df['Diabetes_012']

print("Undersampled class distribution in train:")
print(y_train_balanced.value_counts())
```
